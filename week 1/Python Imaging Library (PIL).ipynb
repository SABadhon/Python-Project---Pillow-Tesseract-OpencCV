```python
image = Image.open(file).convert('RGB')
```

```python
from PIL import ImageEnhance
# starts from 0.0 which is dark and goes till 1.0

enhancer = ImageEnhance.Brightness(image)
images = []
for i in range(0,10):
    images.append(enhancer.enhance(i/10))
    
print(images)
```

```python
help(PIL.Image.new)
```

```python
# tiling the images in one image
first_image = images[0]
from PIL import Image
contact_sheet = PIL.Image.new(first_image.mode,
                              (first_image.width, 10*first_image.height))

current_location = 0
for img in images:
    contact_sheet.paste(img, (0,current_location))
    current_location = current_location + 450

contact_sheet = contact_sheet.resize((160,900))
display(contact_sheet)
```

```python
canvus = PIL.Image.new(first_image.mode,
                      (first_image.width*3, first_image.height*3))
x = 0
y = 0

for img in images[1: ]:
    canvus.paste(img, (x,y) )
    if x + first_image.width == canvus.width:
        x = 0
        y = y + first_image.height
    else:
        x = x+ first_image.width

canvus = canvus.resize(( int(canvus.width/2), int(canvus.height/2) ))
display(canvus)
```
