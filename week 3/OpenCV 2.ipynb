```python
import cv2
face_casc = cv2.CascadeClassifier('readonly/haarcascade_frontalface_default.xml')
eye_casc = cv2.CascadeClassifier('readonly/haarcascade_eye.xml')
```

```python
img = cv2.imread('readonly/floyd.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces = face_casc.detectMultiScale(gray)
faces
```

```python
faces.tolist()[0]
```

```python
import PIL
from PIL import Image

image = Image.fromarray(gray, 'L')

from PIL import ImageDraw
rec = faces.tolist()[0]

drawing = ImageDraw.Draw(image)
drawing.rectangle(rec, fill = None, outline= 'white')

display(image)
```

```python
image = Image.fromarray(gray, 'L')
drawing = ImageDraw.Draw(image)
drawing.rectangle( (rec[0], rec[1], rec[0]+rec[2], rec[1]+rec[3]), fill = None, outline = 'white' )
display(image)
```

```python
def show_faces(faces):
    pil_img = Image.open('readonly/msi_recruitment.gif').convert('RGB')
    drawing = ImageDraw.Draw(pil_img)
    for x,y,w,h in faces:
        drawing.rectangle( (x,y,x+w, y+h), fill=None, outline = 'blue' )
    display(pil_img)
```

```python
img = cv2.imread('recruitment.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv_bin_img = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)[1]
faces = face_casc.detectMultiScale(cv_bin_img)
show_faces(faces)
```

```python
faces = face_casc.detectMultiScale(img, 1.05)
show_faces(faces)

faces = face_casc.detectMultiScale(img, 1.15)
show_faces(faces)
```
