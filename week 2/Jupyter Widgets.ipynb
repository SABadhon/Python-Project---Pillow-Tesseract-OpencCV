```python
path = 'readonly/storefront.jpg'
```

```python
from PIL import Image
image = Image.open(path)
display(image)
```

```python
from ipywidgets import interact
from PIL import ImageDraw
```

```python
@interact(left = 100, top=100, right = 200, bottom = 200)

def draw_border(left, top, right, bottom):
    img = image.copy()
    drawing_obj = ImageDraw.Draw(img)
    drawing_obj.rectangle((left, top, right, bottom), fill = None, outline = 'red')
    display(img)
```
